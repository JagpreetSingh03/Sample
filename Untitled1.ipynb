```python
import polars as pl
gapminder=pl.read_csv("https://raw.githubusercontent.com/damien-dupre/data/refs/heads/master/gapminder.csv")
```

```python
from plotnine import *
```

```python
(ggplot(data=gapminder)
+ aes(x="continent", y="lifeExp",color="continent")
+ geom_boxplot()
)
```

```python
(ggplot(gapminder)
  + aes(x="year", y="pop", fill="continent")
  + geom_col()
)
```

```python
(ggplot(gapminder)
  + aes(x="year", y="lifeExp", color="continent")
  + geom_point()
  + geom_smooth(method="loess")
  + facet_wrap("continent", nrow=1)
  + theme_classic()
)
```

```python
(ggplot(gapminder)
  + aes(x="year", y="lifeExp", color="continent")
  + geom_point()
  + facet_wrap("continent", nrow=1)
  + theme_classic()
)
```

```python
import geopandas as gp
```

```python
shape_world = gp.read_file("https://public.opendatasoft.com/api/explore/v2.1/catalog/datasets/world-administrative-boundaries/exports/shp")
```

```python
(ggplot()
  + geom_map(shape_world)
  + coord_fixed ()
)
```

```python
(ggplot(gapminder)
  + geom_boxplot()
  + facet_wrap("Continent")
  + labs(x="year (from 1952 to 2007)", y="Life Expectancy", title="Evolution of Life")
  + theme_classic()
)
```

```python
(ggplot(gapminder)
  + aes(x="continent", y="LifeExp", color="continent")
  + geom_boxplot()
  + facet_wrap("year")
  + labs(x="Continent", y="Life Expectancy", title="Life Expentancy Rate")
  + theme_classic()
  )
```
